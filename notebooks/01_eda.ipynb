# MovieLens 100k — Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA = '../data/raw'
ratings = pd.read_csv(f'{DATA}/ratings.csv')
movies  = pd.read_csv(f'{DATA}/movies.csv')
links   = pd.read_csv(f'{DATA}/links.csv')

print(f'Ratings : {len(ratings):,}')
print(f'Users   : {ratings["userId"].nunique()}')
print(f'Movies  : {ratings["movieId"].nunique()}')
ratings.head()
```

```python
# Rating distribution
ratings['rating'].value_counts().sort_index().plot(kind='bar', color='steelblue')
plt.title('Rating Distribution')
plt.xlabel('Rating'); plt.ylabel('Count')
plt.tight_layout(); plt.show()
```

```python
# Ratings per user vs ratings per movie
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

ratings.groupby('userId').size().hist(bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Ratings per User'); axes[0].set_xlabel('# ratings')

ratings.groupby('movieId').size().hist(bins=50, ax=axes[1], color='salmon')
axes[1].set_title('Ratings per Movie'); axes[1].set_xlabel('# ratings')

plt.tight_layout(); plt.show()
```

```python
# Matrix sparsity
n_users = ratings['userId'].nunique()
n_movies = ratings['movieId'].nunique()
sparsity = 1 - len(ratings) / (n_users * n_movies)
print(f'Matrix size : {n_users} x {n_movies} = {n_users * n_movies:,} possible entries')
print(f'Filled      : {len(ratings):,}')
print(f'Sparsity    : {sparsity:.2%}')
```

```python
# Top 20 most-rated movies
top = (
    ratings.groupby('movieId')
    .agg(count=('rating','count'), mean=('rating','mean'))
    .merge(movies, on='movieId')
    .sort_values('count', ascending=False)
    .head(20)
)
top[['title', 'count', 'mean']].style.format({'mean': '{:.2f}'})
```
